```python
# Bivariate analysis
```

```python
# Age, Cost
plt.scatter(insurance.age, insurance.charges, alpha=0.2)
plt.title("Age vs Insurance Cost")
plt.xlabel("Age (Years)")
plt.ylabel("Cost (USD)")
```

```python
# BMI, Cost
plt.scatter(insurance.bmi, insurance.charges, alpha=0.2)
plt.title("BMI vs Insurance Cost")
plt.xlabel("BMI")
plt.ylabel("Cost (USD)")
```

```python
# Children, Cost
sns.stripplot(x=insurance["children"], y=insurance["charges"], jitter=True, alpha=0.3)
plt.title("Children vs Insurance Cost")
plt.xlabel("Number of Children")
plt.ylabel("Cost (USD)")
```

```python
sns.boxplot(x=insurance["children"], y=insurance["charges"])
plt.title("Children vs Insurance Cost")
plt.xlabel("Number of Children")
plt.ylabel("Cost (USD)")
```

```python
# Smoking, Cost
sns.boxplot(x=insurance["smoker"], y=insurance["charges"])
plt.title("Smoking vs Insurance Cost")
plt.xlabel("Smoker")
plt.ylabel("Cost (USD)")
```

```python
# Region, Cost
sns.stripplot(x=insurance["region"], y=insurance["charges"], jitter=True, alpha=0.3)
plt.title("Region vs Insurance Cost")
plt.xlabel("Region")
plt.ylabel("Cost (USD)")
```

```python
sns.boxplot(x=insurance["region"], y=insurance["charges"])
plt.title("Region vs Insurance Cost")
plt.xlabel("Region")
plt.ylabel("Cost (USD)")
```
